```python
import numpy as np
```

```python
data = np.random.randint(0,100,25)
data
```

```python
np.sort(data)[::-1]
```

```python
def top_k(data, k=10):
  descending_data = np.sort(data)[::-1]
  return descending_data[:k]
```

```python
def top_p(data, p=80):
  new = [d for d in data if d > p]
  return new
```

```python
top_k(data, k =10)
```

```python
top_p(data, p =80)
```
